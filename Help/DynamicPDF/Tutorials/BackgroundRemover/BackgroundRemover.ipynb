(<a href='#Table of Contents'>Top</a>)<a id='Overview'></a><h3>Overview</h3>

The purpose background remover is to obtain the dynamic pair distribution function (PDF, $G(r,E)$) after removal of the multi-phonon background with a fitting procedure.

<center><a href="files/supporting/overview.png"><img src="files/supporting/overview.png" width="600" height="250" alt="supporting/overview.png"></a></center>

The interface is accessible from MantidPlot:

<h2>A Fitting Walkthrough</h2>
This walkthrough carries the user from loading the slices to obtaining the PDF.

(<a href='#Walkthrough'>Top</a>)<a id='Walkthrough.loading'></a><h3>Loading slices</h3>  

We start opening the Slice Selector widget clicking the <code>Load</code> button

<center><a href="files/supporting/click_load.png"><img src="files/supporting/click_load.png" width="300" height="250" alt="supporting/click_load.png"></a></center>

This is the Slice Selector widget prior to loading any data:

<center><a href="files/supporting/sliceSelector_slide1.png"><img src="files/supporting/sliceSelector_slide1.png" width="400" height="750" alt="supporting/sliceSelector_slide1.png"></a></center>

Data that can be loaded is the output from algorithm
[DPDFreduction](http://docs.mantidproject.org/nightly/algorithms/DPDFreduction-v1.html), the structure factor $S(Q,E)$. Usually it is saved as a nexus file. In this walkthrough we will use file [slices_QE.nxs](files/supporting/slices_QE.nxs) describing neutron scattering of liquid aluminum.

To load the file, use the <code>Browse</code> button to search for the file. Once the path is loaded, click in the <code>Load</code> button.

<center><a href="files/supporting/sliceSelector_slide2.png"><img src="files/supporting/sliceSelector_slide2.png" width="400" height="750" alt="supporting/sliceSelector_slide2.png"></a></center>

(<a href='#Walkthrough'>Top</a>)<a id='Walkthrough.fitting'></a><h3>Removing the background of one slice</h3>

After selecting slice 49, we click the <code>Fit</code> button to load the slice in the Background remover:

<center><a href="files/supporting/backgroundRemover_1.png"><img src="files/supporting/backgroundRemover_1.png" width="400" height="150" alt="supporting/backgroundRemover_1.png"></a></center>

There are three built-in fit functions accessible from the <code>Built-in</code> pull-down menu:

<center><a href="files/supporting/backgroundRemover_2.png"><img src="files/supporting/backgroundRemover_2.png" width="200" height="150" alt="supporting/backgroundRemover_2.png"></a></center>

These are:

* <code>Gaussian + LB</code>: $Height \cdot e^{-0.5((Q-PeakCentre)/Sigma)^2} + (A_0+A_1Q)$
* <code>QuadXGauss + LB</code>: $(A_0+A_1Q+A_2Q^2) \cdot Height \cdot e^{-0.5((Q-PeakCentre)/Sigma)^2} + (A'_0+A'_1Q)$
* <code>Quadratic</code>: $A_0+A_1Q+A_2Q^2$

We choose the second funtion. The fitting parameters are then shown in the <i>Model Builder</i>. 

<center><a href="files/supporting/backgroundRemover_3.png"><img src="files/supporting/backgroundRemover_3.png" width="700" height="300" alt="supporting/backgroundRemover_3.png"></a></center>

The Model Builder allows setting bounds and even ties on the fitting parameters. For instance, parameter $Height$ is redundant since it could be included in the quadratic expression. Thus, we fix $Height$ to unity. First we hover the mouse over the value next to <code>Height</code> and right-click to edit the value, then enter 1.

You will notice that a few additional events took place upon fixing $Height$. When manipulating the Model Builder, the interface interprets the user is guessing a fit. After setting $Height=1$ our model reads

<center>$(0+0Q+0Q^2) \cdot 1 \cdot e^{-((Q-0)/0)^2} + (0+0Q)$</code>

which is obviously zero for any value of Q. Evaluation of this model is the flat red line shown in the *background remover* window:

<center><a href="files/supporting/backgroundRemover_8.png"><img src="files/supporting/backgroundRemover_8.png" width="400" height="300" alt="supporting/backgroundRemover_8.png"></a></center>

Below this window is the *residuals* window, which is just the experimental black curve minus the fit red "curve". In this case the residuals is just the experimental curve since our fit is just a flat line with zero intensity

<center><a href="files/supporting/backgroundRemover_9.png"><img src="files/supporting/backgroundRemover_9.png" width="400" height="300" alt="supporting/backgroundRemover_9.png"></a></center>

Finally, under the residuals window is the *PDF* window, which is the result of applying algorithm [PDFFourierTransform](http://docs.mantidproject.org/nightly/algorithms/PDFFourierTransform-v1.html?highlight=pdffouriertransform) to the residuals curve. The options passed onto the algorithm are to the left in the *PDF Options* panel

<center><a href="files/supporting/backgroundRemover_10.png"><img src="files/supporting/backgroundRemover_10.png" width="400" height="300" alt="supporting/backgroundRemover_10.png"></a></center>

Thus, every time the model is changed in the *Model Builder*, a new guess is plotted, then residuals are calculated, and then a PDF is obtained. We will use these features to obtain a good initial guess prior to do the fit.

We will first try to fit the slice to a simple gaussian with a flat background. A plausible initial model for the simple Gaussian would be would be  
<code>$(2.5+0Q+0Q^2) \cdot 1 \cdot e^{-((Q-13)/4)^2} + (0+0Q)$</code>  
Here the intensity of the gaussian is $A_0=2.5$, centered at $PeakCentre=13$ with standard deviation $Sigma=4$.

<center><a href="files/supporting/backgroundRemover_11.png"><img src="files/supporting/backgroundRemover_11.png" width="700" height="300" alt="supporting/backgroundRemover_11.png"></a></center>

We fix parameters $A_1=0$, $A_2=0$ in the quadratic expresion and also the slope in the background as we did before for parameter $Height$:

<center><a href="files/supporting/backgroundRemover_12.png"><img src="files/supporting/backgroundRemover_12.png" width="200" height="200" alt="supporting/backgroundRemover_12.png"></a></center>

Then press the <code>Fit</code> to optimize the model. The result:

<center><a href="files/supporting/backgroundRemover_13.png"><img src="files/supporting/backgroundRemover_13.png" width="700" height="200" alt="supporting/backgroundRemover_13.png"></a></center>

Next we free up the background slope. Hover the mouse over parameter $A_1$ in the background, left-click and then <code>Remove tie</code>

<center><a href="files/supporting/backgroundRemover_14.png"><img src="files/supporting/backgroundRemover_14.png" width="200" height="200" alt="supporting/backgroundRemover_14.png"></a></center>

The model is now a Gaussian plus linear background. Press <code>Fit</code> for a new optimization.

<center><a href="files/supporting/backgroundRemover_15.png"><img src="files/supporting/backgroundRemover_15.png" width="700" height="200" alt="supporting/backgroundRemover_15.png"></a></center>

Finally, we remove the ties in the quadratic expression and again optimize:

<center><a href="files/supporting/backgroundRemover_16.png"><img src="files/supporting/backgroundRemover_16.png" width="700" height="200" alt="supporting/backgroundRemover_16.png"></a></center>

The corresponding PDF shows the main peak in the liquid Aluminum at about 3 Angstroms:
